```python
titanic = pd.read_csv('titanic.csv')
titanic = titanic[['Survived', 'Pclass', 'Name', 'Sex', 'Age', 'SibSp',
       'Parch', 'Fare']]
titanic.rename(columns = {'SibSp': 'Siblings/Spouses Abroad', 'Parch': 'Parents/Child Abroad'}, inplace='True')
titanic
```

```python
data_to_add = {'Survived':[1,0], 'Pclass':[1,3], 'Name':['Miss. Rose Dewitt Bukater', 'Mr. Jack Dawson'], 'Sex': ['female', 'male'], 'Age': [17,20], 'Siblings/Spouses Abroad':[2,0],
       'Parents/Child Abroad':[1,0], 'Fare':[25,30]}

addendum = pd.DataFrame(columns=titanic.columns, data = data_to_add)
addendum 
```

```python
pd.concat([titanic,addendum]).reset_index(drop=True)
```

```python
data_to_add = {'Pclass':[1,2,3], 'Total Cabin Amount':[371, 105, 84] }

extra_info = pd.DataFrame(columns=['Pclass', 'Total Cabin Amount'], data = data_to_add)

extra_info
```
