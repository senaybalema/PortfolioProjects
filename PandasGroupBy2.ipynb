```python
df = pd.DataFrame(
    {
        "Publish date": [
            pd.Timestamp("2000-01-02"),
            pd.Timestamp("2000-01-02"),
            pd.Timestamp("2000-01-09"),
            pd.Timestamp("2000-01-16")
        ],
        "ID":[0,1,2,3],
        "Price":[10,20,30,40]
    }
)
```

```python
df
```

```python
df.groupby('Publish date')['Price'].mean()
```

```python
df.groupby(pd.Grouper(key ='Publish date', freq='1W'))['Price'].mean()
```

```python
df.groupby(pd.Grouper(key ='Publish date', freq='2W'))['Price'].mean()
```

```python
df.groupby(pd.Grouper(key ='Publish date', freq='1D'))['Price'].mean()
```

```python
df.groupby(pd.Grouper(key ='Publish date', freq='1M'))['Price'].mean()
```

```python
df.groupby(pd.Grouper(key ='Publish date', freq='1D'))['Price'].mean().fillna(0)
```
