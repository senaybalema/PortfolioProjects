```python
## Consolidated Strategy for Targeting 

cm = sns.light_palette("green", as_cmap=True)
## Use a green map
correlation_analysis = pd.DataFrame(data[['Amount Collected','Campaign (Email)','Campaign (Flyer)','Campaign (Phone)','Sales Contact 1',
                                          'Sales Contact 2','Sales Contact 3','Sales Contact 4','Sales Contact 5']].corr()['Amount Collected']).reset_index()
## We use the corrletion function and select important columns for our correlation pertaining to amount collected

correlation_analysis.columns=['Impacting Variable', 'Degree of Linear Impact (Correlation)']
# Change the name of the columns 
correlation_analysis= correlation_analysis[correlation_analysis['Impacting Variable'] != 'Amount Collected']
## WE remove amount collected because it has 100 % correlation
correlation_analysis= correlation_analysis.sort_values('Degree of Linear Impact (Correlation)', ascending=False)
#Sort our correlation values by descending order amd set pur value to 2 deciamals 
correlation_analysis.style.background_gradient(cmap=cm).set_precision(2)
```
