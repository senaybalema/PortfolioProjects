```python
#transform
```

```python
dftransform = sns.load_dataset('tips')
```

```python
dftransform
```

```python
dftransform['medianbill'] = df.groupby('time')['total_bill'].transform('median')
dftransform
```

```python
dftransform.query('time=="Lunch"')
```

```python
dftransform.query('time=="Dinner"')
```

```python
dftransform.query('total_bill > medianbill')
```

```python
dftransform[dftransform['total_bill'] > dftransform['medianbill']]
```

```python
#show a lambda example --> z-score for each group:Where data lies on a normal distribution curve
```

```python
dftransform['zscore']= df.groupby('time')['total_bill'].transform(lambda x: (x - x.mean()) / x.std())
```

```python
dftransform.head()
```

```python
#apply
```

```python
dftransform.groupby('time').apply(lambda x: (100*  x['tip']/x['total_bill']).mean()).to_frame('avgtippercent').reset_index()
```

```python
#Question 1 --> Interquartile Range Q3 - Q1 for Tips with boroughs --> Apply
```

```python
tx.query('pickup_borough=="Bronx"')
```

```python
#Question 2 -Show Mean, Min, Max Fare for Payment Type and Taxi Color
#relabel Mean as avg, min as minimum, max as maximum
```

```python
tx.groupby(['payment', 'color'])['fare'].agg(avg='mean', minimum='min', maximum = 'max')
```

```python
tx.groupby(['payment', 'color'])['fare'].agg(['mean','min','max']).rename(columns={'min':'minimum', 'max':'maximum'})
```

```python
#Question 3a Output the avgtotal for taxi color total to each raw
```

```python
tx['avgtotal'] = tx.groupby('color')['total'].transform('mean')
```

```python
tx.head()
```

```python
#Question 3b Output the rows that have a total greater then the avg color 
```

```python
tx.query('total > avgtotal')
```
