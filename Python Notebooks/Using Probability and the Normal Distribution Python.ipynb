# Normal Distribution and Probability 
## Feature of a Normal Distribution 
- Bell shaped
- Y axis equals the relative probabilities 
- Rare events are the extreme ends of the curve 
- The width of the curve is defined by the standard deviation 
- 68% of all the data falls between 1 and standard deviation from the mean
- 95 between standard deviations 
- 99.7% with in 3 standard deviations 
- Standard deviations is going to determine the width of the curve

```python
battery_sample = norm.rvs(loc=6, scale=1.5, size=1000)

bat_mean = battery_sample.mean()

bat_std = battery_sample.std()



plt.axvline(bat_mean)

plt.axvline(bat_mean + bat_std,ls= "--",color = 'black')

plt.axvline(bat_mean - bat_std, ls= "--",color = 'black')

sns.kdeplot(norm.rvs(loc=6, scale=1.5, size=1000), fill=True)
```

```python
norm.interval(.68, loc=6, scale=1.5)
```

#### Between 4 and 8 hours

```python
below_4_hours = norm.cdf(4,loc=6,scale=bat_std)
below_8_hours = norm.cdf(8,loc=6,scale=bat_std)
print(below_8_hours-below_4_hours)
```
