```python
#import libraries 

import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
# load in the data

data = pd.read_excel(r"C:\Users\senay\Downloads\Online Retail.xlsx")
                     
data.info()
```

```python
#take a look at the head
data.head()
```

```python
# drop rows with no customer id 

data = data.dropna(subset=['CustomerID'])
data.info()
```

```python
#create an invoice month from invoice date
# import datetime libraries
import datetime as dt
#create function for month 
def get_month(x):
    return dt.datetime(x.year, x.month, 1)

#apply the function to invoice date
data['InvoiceMonth'] = data['InvoiceDate'].apply(get_month)
data.tail()
```

```python
#create a column index with the minimum invoice date aka first time customer was aquired 

data['Cohort Month'] = data.groupby('CustomerID')['InvoiceMonth'].transform('min')

data.tail(30)
```

```python
#Create a date element function to get a series for subtraction 

def get_date_elements(df, column): 
    day = df[column].dt.day
    month = df[column].dt.month
    year = df[column].dt.year 
    return day,month,year 
    
```

```python
#get date elements for our cohort and invoice columns 
_,Invoice_month,Invoice_Year = get_date_elements(data, 'InvoiceMonth')
_,Cohort_month,Cohort_Year = get_date_elements(data, 'Cohort Month')
```

```python
#check the series 
Invoice_month[:10]
```

```python
#check the series 
Cohort_Year[:10]
```

```python
#Create a cohort index, lowest value is 1 since the first month will be 1 rather than zero

year_diff = Invoice_Year - Cohort_Year
month_diff = Invoice_month - Cohort_month

data['Cohort Index'] = year_diff*12+month_diff+1
data.head()
```

```python
data.tail()
```

```python
#Create a pivot table 
#Count the customer ID by grouping by Cohort Month and Cohort Index

cohort_data = data.groupby(['Cohort Month', 'Cohort Index'])['CustomerID'].apply(pd.Series.nunique).reset_index()
cohort_data
```

```python
#Create a pivot taable
cohort_table = cohort_data.pivot(index = 'Cohort Month', columns = ['Cohort Index'], values = 'CustomerID')
cohort_table 
```

```python
#change index from date to string for heatmap

cohort_table.index = cohort_table.index.strftime('%B %Y')
```

```python
#Visualize our results in a heat map
plt.figure(figsize=(21,10))
sns.heatmap(cohort_table, annot=True, cmap='Blues')
```

```python
#Cohort Table for percentage
#Divde the cohort table by the first element in each row 
new_cohort_table = cohort_table.divide(cohort_table.iloc[:,0],axis = 0)
new_cohort_table
```

```python
#Create our percentage visual heat map

plt.figure(figsize=(21,10))
sns.heatmap(new_cohort_table, annot=True, cmap='magma', fmt='.0%')
```

```python
#Experimenting heat map colors 

plt.figure(figsize=(21,10))
sns.heatmap(new_cohort_table, annot=True, cmap='coolwarm', fmt='.0%')
```
